```python
def var_summary(x):
    return pd.Series(
        [x.count(), 
        x.isnull().sum(),
        x.sum(),
        x.mean(),
        x.median(),
        x.std(),
        x.var(),
        x.min(),
        x.quantile(0.01),
        x.quantile(0.05),
        x.quantile(0.10),
        x.quantile(0.25),
        x.quantile(0.50),
        x.quantile(0.75),
        x.quantile(0.90),
        x.quantile(0.95),
        x.quantile(0.99),
        x.max()],
        index=['N',
               'NMISS',
               'SUM',
               'MEAN', 
               'MEDIAN', 
               'STD',
               'VAR', 
               'MIN', 
               'P1',
               'P5',
               'P10',
               'P25/Q1',
               'P50/Q2',
               'P75/Q3',
               'P90',
               'P95', 
               'P99',
               'MAX'])
```

```python
advt.apply(lambda x: var_summary(x)).T
```

```python
# EDA with autoviz
from autoviz.AutoViz_Class import AutoViz_Class
advt = pd.read_csv("Advertising.csv")
print("Dataset shape:", advt.shape)
print(advt.head())

print("\n=== DATA PROFILING ===")
print(advt.info())
print("\n=== STATISTICAL SUMMARY ===")
print(advt.describe())
print("\n=== MISSING VALUES ===")
print(advt.isnull().sum())
print("\n=== DATA TYPES ===")
print(advt.dtypes)
AV = AutoViz_Class()
detailed_analysis = AV.AutoViz(
    filename="",
    dfte=advt,
    depVar="Sales", 
    verbose=2,  
    lowess=True, 
    chart_format="png"
)
```
